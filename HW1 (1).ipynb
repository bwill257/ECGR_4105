```python
#Brandon Williams
#ECGR 4105 Homework 1

#Import Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

#Importing Data from D3.csv
url = 'https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/b4cf4ac187d647a443dc011c4ee85fbb42e26096/Dataset/D3.csv'
df = pd.read_csv(url)
print(df.head())

#Seperating out the X1, X2, X3, and Y values from the D3.csv
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
Y = df.values[:,3]

#Training examples
Tx1 = len(X1)
Tx2 = len(X2)
Tx3 = len(X3)
TY = len(Y)

m = len(Y)  # Number of training examples
n = len(X1)  # Number of training examples

#Display first 5 records and the total number of training examples
print('X1 = ', X1[: 5])
print('X2 = ', X2[: 5])
print('X3 = ', X3[: 5])
print('Y = ', Y[: 5])
print('m = ', m)
print('n = ', n)

# Function definitions
def gradient_descent(X, y, theta, alpha, trials):
    m = len(y)  # Number of training examples
    cost_history = np.zeros(trials)

    for i in range(trials):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J
```

```python
#Linear Regression for X1
T1 = np.ones((Tx1, 1))
T2 = X1.reshape(Tx1, 1)
X = np.hstack((T1, T2))
theta = np.zeros(2)
theta = [0., 0.]
trials = 1800
alph = 0.05
theta, cost_history = gradient_descent(X, Y, theta, alph, trials)
print('X1:')
print('Value of Theta = ', theta)
print('Value of Cost = ', cost_history)
print(f"Linear model: Y = {theta[0]} + {theta[1]}*X1")

#Parameters for Iteration Plot of X1
plt.scatter(X[:, 1], Y, color='blue', marker='+', label='Training Data')
plt.plot(X[:, 1], X.dot(theta), color='black', label='Linear Regression')
plt.rcParams["figure.figsize"] = (15, 8)
plt.grid(True)
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Regression Best Fit') #Title
plt.legend()
plt.show() #Displaying the Plot
```

```python
#Parameters for Loss Plot for X1
plt.plot(range(1, trials + 1), cost_history, color='red')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of Iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of Gradient Descent')
plt.show()
```
